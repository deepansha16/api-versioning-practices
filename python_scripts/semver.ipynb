```python
ap=pd.read_csv('final_df.csv')
```

```python
import plotly.express as px
import plotly.graph_objects as go

data2= ap.query("api_spec_id==641").sort_values(by='commit_date')
data2['dupe'] = data2.info_version.where(~data2.info_version.duplicated(), '')
data2['label'].replace({np.nan: ''}, inplace=True)
data2['dupe1'] = data2.label.where(~data2.label.duplicated(), '').replace({np.nan: ''})

import plotly.graph_objects as go
fig = go.Figure()

fig = px.scatter(data2, x='commit_date', y='API_changes', color='info_version', title='Ingest API Version Upgrades', hover_name='info_version', hover_data=['info_version','API_changes'])
fig.add_trace(go.Scatter(mode='lines',
                         x=data2["commit_date"],
                         y=data2["API_changes"],
                         line_color='black',
                         line_width=0.6,
                         line_shape='hv',
                         showlegend=False
                       )
             )

for _,row in data2.iterrows():
    fig.add_annotation(
        go.layout.Annotation(
            x=row["commit_date"],
            y=row["API_changes"],
            text=row['dupe'],
            showarrow=False,
            align='center',
            yanchor='auto',
            yshift=14,
            textangle=-0)
       )
    fig.add_annotation(
        go.layout.Annotation(
            x=row["commit_date"],
            y=row["API_changes"],
            text=row['label'],
            showarrow=False,
            align='center',
            yanchor='auto',
            yshift=-10,
            textangle=-0)
         )

 
fig.update_layout( template='plotly_white',height=700, width=1600, xaxis_title="Date", yaxis_title="API Changes", font=dict(size=12))
fig.update_traces(marker_size=7, marker_line_width=1, marker_line_color='black', showlegend=False, textposition='bottom center')

fig.show("svg")
```
